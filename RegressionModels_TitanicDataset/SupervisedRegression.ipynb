```python
import pandas as pd
import numpy as np 
```

## Titanic Veri Setini Yükleme ve İlk 5 Satırı Görüntüleme

```python
df = pd.read_csv('Titanic/train.csv')
df.head()
```

## Veri Setindeki Eksik Değerlerin Sayısını Gösterme

```python
df.isnull().sum()
```

## Eksik Değerleri Ortalama ve Mod ile Doldurma

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

## Veri Setindeki Eksik Değerlerin Sayısını Gösterme

```python
df.isnull().sum()
```

## İstenmeyen Sütunları Düşürme ve Hedef Değeri Ayarlama

```python
X = df.drop(['Name' , 'Ticket' , 'Cabin'], axis =1)
y = df['Survived']
```

## Veri Setinin Genel Bilgilerini Görüntüleme

```python
X.info()
```

## Kategorik Veriyi Sayısal Veriye Dönüştürme (Label Encoding)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['Sex'] = le.fit_transform(X['Sex'])
```

## Kategorik Veriyi One-Hot Encoding ile Sayısal Veriye Dönüştürme

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder(sparse_output=False)
embarked_encoded = one.fit_transform(X[['Embarked']])
embarked_cols = one.get_feature_names_out(['Embarked'])
embarked_df = pd.DataFrame(embarked_encoded, columns=embarked_cols, index=X.index)
X = X.drop('Embarked', axis=1)
X = pd.concat([X, embarked_df], axis=1)
```

## Veri Setinin İlk 5 Satırını Görüntüleme

```python
X.head()
```

## Veri Setini Eğitim ve Test Setlerine Ayırma

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.3)
```

## Veri Setini Ölçeklendirme (StandardScaler)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Doğrusal Regresyon Modeli ile Eğitim Yapma

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

## Test Verisi ile Tahmin Yapma

```python
y_pred = model.predict(X_test_scaled)
```

## Model Performansını Değerlendirme (MAE, MSE, R2)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)

print(f"MAE: {mae} MSE: {mse} R2: {r2}")
```

## Karar Ağacı Regresyon Modeli ile Eğitim Yapma

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=5, random_state=42)
model.fit(X_train_scaled, y_train)
```

## Test Verisi ile Tahmin Yapma

```python
y_pred = model.predict(X_test_scaled)
```

## Model Performansını Değerlendirme (MAE, MSE, R2)

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)

print(f"MAE: {mae} MSE: {mse} R2: {r2}")
```

## Random Forest Modeli ile Eğitim Yapma

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=5, max_depth=5, random_state=42)
model.fit(X_train_scaled, y_train)
```

## Test Verisi ile Tahmin Yapma

```python
y_pred = model.predict(X_test_scaled)
```

## Model Performansını Değerlendirme (MAE, MSE, R2)

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)

print(f"MAE: {mae} MSE: {mse} R2: {r2}")
```

## Modeli GridSearchCV ile Hiperparametre Optimizasyonu

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": list(range(5,15)),
    "max_depth": list(range(1,10))
}

grid = GridSearchCV(  RandomForestRegressor(), param_grid, scoring="r2", cv=10 )
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_estimator_, grid.best_score_, grid.best_params_
```
