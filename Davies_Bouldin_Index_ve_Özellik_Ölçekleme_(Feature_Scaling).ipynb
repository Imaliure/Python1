# Davies-Bouldin Index ve Özellik Ölçekleme (Feature Scaling)

Bu notebook, kümeleme değerlendirme metriği olan Davies-Bouldin Index (DBI) ve özellik ölçekleme (scaling) kavramlarını açıklamakta ve Python örnekleriyle uygulamaktadır.


## Davies-Bouldin Index Nedir?

Davies-Bouldin Index, denetimsiz öğrenmede kümeleme kalitesini değerlendirmek için kullanılır.
Düşük DBI değeri, iyi bir kümeleme performansını ifade eder.


## Veri Oluşturma ve Ölçekleme

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

# Örnek veri seti
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Scaling işlemi
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## K-Means ve DBI Hesaplama

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score

# K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Davies-Bouldin Index
dbi = davies_bouldin_score(X_scaled, labels)
print("Davies-Bouldin Index:", dbi)
```

**DBI (Davies-Bouldin Index) skorunun 0.28 çıkması, çok iyi bir kümeleme sonucu elde ettiğini gösterir.**

## K-Means Kümeleme Görselleştirmesi (Davies-Bouldin Index ile)

```python
import matplotlib.pyplot as plt

# Kümeleme sonrası verileri renklendirerek görselleştirelim
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', s=50)
plt.title("K-Means Kümeleme Sonucu (DBI: {:.2f})".format(dbi))
plt.xlabel("Özellik 1 (scaled)")
plt.ylabel("Özellik 2 (scaled)")
plt.grid(True)
plt.show()
```

# Özellik Ölçekleme (Scaling) Nedir?

Verilerin aynı ölçekte olması, özellikle mesafe tabanlı algoritmalar için önemlidir.
Kullanılan başlıca yöntemler:
- Min-Max Scaling
- Z-Score Standardizasyonu (StandardScaler)

# Scaling Örneği

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_minmax = scaler.fit_transform(X)
print(f'Olceklenmemis veri :\n {X[:5]}')
print(f'Olceklenmis veri :\n {X_minmax[:5]}')
```
