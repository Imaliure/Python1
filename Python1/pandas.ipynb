```python
import pandas as pd
import numpy as np
```

**1. df.head(n) – İlk n Satırı Getirir**

Açıklama: DataFrame'in ilk n satırını döndürür. Varsayılan olarak ilk 5 satırı gösterir.

```python
df = pd.DataFrame({'A':[1,2,3,np.NAN,4,5],'B':[6,7,8,9,np.NAN,10]})
df.head(3)
```

**2. df.tail(n) – Son n Satırı Getirir**

Açıklama: DataFrame'in son n satırını döndürür.

```python
df.tail(2)
```

**3. df.info() – Veri Kümesinin Genel Bilgilerini Gösterir**

Açıklama: DataFrame'in sütun bilgilerini, veri tiplerini ve eksik değerleri gösterir.

```python
df.info()
```

**4. df.describe() – İstatistiksel Özet Verir**

Açıklama: Sayısal sütunlar için temel istatistikleri (ortalama, min, max, std, vb.) döndürür.

```python
df.describe()
```

**5. df.isnull().sum() – Eksik Değerleri Sayar**

Açıklama: Her sütundaki eksik (NaN) değerlerin sayısını döndürür.

```python
df.isnull().sum()
```

**6. df.fillna(value) – Eksik Değerleri Doldurur**

Açıklama: NaN değerlerini belirli bir değerle değiştirir.

```python
df.fillna(0,inplace=True)
df
```

**7. df.dropna() – Eksik Değer İçeren Satırları veya Sütunları Siler**

Açıklama: NaN içeren satırları (axis=0) veya sütunları (axis=1) siler.

```python
df.dropna(inplace=True)
df
```

**8. df.rename(columns={"EskiAd": "YeniAd"}) – Sütun İsimlerini Değiştirir**


Açıklama: DataFrame içindeki sütun adlarını değiştirmek için kullanılır.

```python
df.rename(columns={"A": "X", "B": "Y"}, inplace=True)
df
```

**9. df.shape – DataFrame’in Boyutunu Döndürür**

Açıklama: DataFrame’in satır ve sütun sayısını gösteren (satır, sütun) şeklinde bir tuple döndürür.

```python
df.shape
```

**10. df.sort_values(by="SütunAdı", ascending=True) – Satırları Sıralar**

Açıklama: Belirtilen sütuna göre DataFrame sıralanır.

```python
df.sort_values(by="X", ascending=False, inplace=True) # "A" sütununa göre azalan sıralama
df
```

```python
df.sort_values(by="X", ascending=True, inplace=True) # "A" sütununa göre artan sıralama
df
```
